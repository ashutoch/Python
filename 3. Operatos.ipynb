```python

# //! Arithmetic Operator

x = 10
y = 3

print(x + y)

print(x - y)

print(x * y )
print(x ** y ) #exponent

print(x / y )
print(x // y ) #floor division

print(x % y )
```

```python

# //! Comparison Operator
# Returns a boolean value

x = 5
y = 10

print (x > y)

print (x < y)

print (x == y)

print (x != y)

print (x >= y)

print (x <= y)
```

```python

# //! Logical Operators

x = True
y = False

print ( x and y )

print ( x or y )

print ( not x )

print ( not y )
```

```python

# //! Membership Operator
# Tests for membership in a sequence

myList = [1, 2, 3, 4, 5]

print(3 in myList) #? Checks if an item is on the list

print(6 not in myList) #? Checks if an item is not on the list
```

```python

# //! Identity Operator
# Compares the memory location of two objects

x = [1, 2, 3]
y = [1, 2, 3]
z = x

print( x is z) #? checks if the tow variables point to the same object

print ( x is not y) #? checks if the two variables do not point to the same object

#? Even tho x & y have the same values they don't point to the same object thats why the identity operation fails

print ( x == y ) # values are same, i.e., same content
print( x is y ) # both are different objects and points to different memeory location
```

```python
# Project 2
# //! Text to speech

import pyttsx3

def speak(text):
    engine = pyttsx3.init('sapi5')
    # engine.setProperty('rate', 150)
    engine.say(text)
    engine.runAndWait()
    engine.stop()

speak("This is line number 1")
speak("This is line number 2")
speak("This is line number 3")
```

```python
# Check if a number is between 10 and 50

num = int(input("Enter a number: "))
if num >= 10 and num <= 50:
    print(f"{num} is between 10 and 50")
else:
    print(f"{num} is not between 10 and 50")
```
